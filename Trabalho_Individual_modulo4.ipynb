# Importação Pandas

```python
import pandas as pd
import numpy as np
```

# Tabela Base


```python
colunas_dic = {'Dia': ["Segunda","Terça","Quarta","Quinta","Sexta","Sabado","Domingo"],
  'Horas Trabalhadas': [6,7,8,6,7,5,4],
  'Bugs Corrigidos': [3,2,1,4,3,2,1],
  'Tarefas Concluídas': [5,4,6,4,5,3,2]}

relatorio_dt= pd.DataFrame(colunas_dic)
relatorio_dt

```

```python
# Correlação
relatorio_dt.corr()
```

# Tabela Horas Trabalhadas

```python
#Cria Tabela copiando duas colunas da Tabela Base
horas_df = relatorio_dt[["Dia","Horas Trabalhadas"]].copy()

#Adiciona 2 Colunas, Soma Acumulada e média Acumulada:
horas_df['Horas Acumuladas'] = horas_df['Horas Trabalhadas'].cumsum()
horas_df['Média Acumulada de Horas'] = round(horas_df['Horas Acumuladas'] / ( horas_df.index + 1),2)
horas_df

```

```python
media = horas_df["Horas Trabalhadas"].mean()
print("Media:", round(media, 2))

desvio = horas_df["Horas Trabalhadas"].std()
print("Desvio Padrão:", round(desvio,2))
print("Porcentagem Desvio/media [%]:", round(desvio/media*100,2) )

limite2= media + desvio
limite1= media - desvio

print("Dias que estão fora do desvio padrão:")
#Imprimir dias que tão fora do dia padrao com LOC
tab_aux = horas_df[["Dia", "Horas Trabalhadas"]].copy()
selecao = tab_aux.loc[(limite2 < horas_df["Horas Trabalhadas"]) | (limite1 > horas_df["Horas Trabalhadas"])]
display(selecao)
```

# Tabela Bugs Corrigidos

```python
#Cria Tabela copiando duas colunas da Tabela Base
bugs_df = relatorio_dt[["Dia",'Horas Trabalhadas',"Bugs Corrigidos"]].copy()

#Adiciona 2 Colunas, Soma Acumulada e média Acumulada:
bugs_df['Bugs Corrigidos Acumulados'] = bugs_df['Bugs Corrigidos'].cumsum()
bugs_df['Média Acumulada'] = round(bugs_df['Bugs Corrigidos Acumulados'] / ( bugs_df.index + 1),2)


#Adiciona 1 COLUNA: BUGS/HORA TRABALHADA
bugs_df['Bugs Corrigidos por hora'] =  round(bugs_df['Bugs Corrigidos']/bugs_df['Horas Trabalhadas'],2)
bugs_df
```

```python
media = bugs_df['Bugs Corrigidos por hora'].mean()
print("Media:", round(media, 2))


desvio = bugs_df['Bugs Corrigidos por hora'].std()
print("Desvio Padrão:", round(desvio,2))

print("Porcentagem Desvio/media [%]:", round(desvio/media*100,2) )

limite2= media + desvio
limite1= media - desvio

print("Dias que estão fora do desvio padrão:")
#Imprimir dias que tão fora do dia padrao com LOC
tab_aux = bugs_df[["Dia", 'Bugs Corrigidos por hora']].copy()
selecao = tab_aux.loc[(limite2 < bugs_df['Bugs Corrigidos por hora']) | (limite1 > bugs_df['Bugs Corrigidos por hora'])]
display(selecao)
```

# Tabela Tarefas Concluídas

```python
#Cria Tabela copiando duas colunas da Tabela Base
tarefas_df = relatorio_dt[["Dia",'Horas Trabalhadas',"Tarefas Concluídas"]].copy()

#Adiciona 2 Colunas, Soma Acumulada e média Acumulada:
tarefas_df['Tarefas Acumuladas'] = tarefas_df['Tarefas Concluídas'].cumsum()
tarefas_df['Média Acumulada de Tarefas'] = round(tarefas_df['Tarefas Acumuladas'] / ( tarefas_df.index + 1),2)



#Adiciona 1 COLUNA: Tarefas/HORA TRABALHADA
tarefas_df['Tarefas Concluídas por hora'] =  round(tarefas_df['Tarefas Concluídas']/tarefas_df['Horas Trabalhadas'],2)
tarefas_df
```

```python
media = tarefas_df['Tarefas Concluídas por hora'].mean()
print("Media:", round(media, 2))

desvio = tarefas_df['Tarefas Concluídas por hora'].std()
print("Desvio Padrão:", round(desvio,2))

print("Porcentagem Desvio/media [%]:", round(desvio/media*100,2) )

limite2= media + desvio
limite1= media - desvio

print("Dias que estão fora do desvio padrão:")
#Imprimir dias que tão fora do dia padrao com LOC
tab_aux = tarefas_df[["Dia", 'Tarefas Concluídas por hora']].copy()
selecao = tab_aux.loc[(limite2 < tarefas_df['Tarefas Concluídas por hora']) | (limite1 > tarefas_df['Tarefas Concluídas por hora'])]
display(selecao)
```

# Eventos (Bugs+Tarefas)

```python
#Cria Tabela copiando duas colunas da Tabela Base
eventos_df = relatorio_dt[["Dia",'Horas Trabalhadas']].copy()
eventos_df["Eventos"] = relatorio_dt["Tarefas Concluídas"]+relatorio_dt["Bugs Corrigidos"]
eventos_df

#Adiciona 1 COLUNA: Eventos/HORA TRABALHADA
eventos_df['Eventos Realizados por hora'] =  round(eventos_df["Eventos"]/eventos_df['Horas Trabalhadas'],2)
eventos_df
```

# Análise

1 - Podemos perceber que há uma alta correlação entre a quantidade de horas trabalhadas e tarefas concluidas, o que indica que a quantidade de horas trabalhadas é afetada majoritariamente pelas quantidade de tarefas. Já a correlação entre bugs corrigidos e horas trabalhadas não é significativa, o que indica que esses nao interferem significavamente no horario empregado.

2 - O a quantidade de bugs varia muito dentro da semana,a razão entre o Desvio padrão e sua media é 47.32 %. Já a de tarefas é de 17,11%, proximo a de horas: 21,9%. Isso indica que a correção de bugs pode variar muito em quantidade durante ao longo da semana, apesar nao afetar tanto a quantidade de horas de trabalho empregadas

3 - Analisando dia com mais horas trabalhadas temos o dia com mais Tarefas concluidas e o segundo com maior média por hora, mas uma quantidade de Eventos(tarefa+bugs) por hora baixa, o que reforça que as tarefas são diretamente proporcionais às horas trabalhadas e os bugs corrigidos pouco afentam as horas trabalhadas

4 - Analisando o dia com menos horas trabahadas temos o menor numero de tarefas concluidas e a menor média por hora, refoçando a ligação entre quantidade de tarefas e horas trabalhadas.
